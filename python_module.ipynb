### Exploring Data

```python
# Univariate Analysis - Distribution plots
numerical_cols = [
    'Minimal price (gross) [PLN]', 'Engine power [KM]', 'Maximum torque [Nm]', 'Battery capacity [kWh]', 
    'Range (WLTP) [km]', 'Wheelbase [cm]', 'Length [cm]', 'Width [cm]', 'Height [cm]', 
    'Minimal empty weight [kg]', 'Permissable gross weight [kg]','Maximum load capacity [kg]', 'Maximum speed [kph]', 
    'Boot capacity (VDA) [l]', 'Acceleration 0-100 kph [s]', 'mean - Energy consumption [kWh/100 km]'
]
```

```python
# Bivariate Analysis - Categorical vs Numerical
# Boxplot of Minimal price (gross) [PLN] by Drive type and Type of brakes
plt.figure(figsize=(12, 6))

# Boxplot for Drive type
plt.subplot(1, 2, 1)
sns.boxplot(x='Drive type', y='Minimal price (gross) [PLN]', data=df, palette='pastel')
plt.title('Price by Drive Type')

# Boxplot for Type of brakes
plt.subplot(1, 2, 2)
sns.boxplot(x='Type of brakes', y='Minimal price (gross) [PLN]', data=df, palette='pastel')
plt.title('Price by Type of Brakes')

plt.tight_layout()
plt.show()
```

```python
# Hypothesis Testing - Comparing means for Electric Range based on Drive Type
from scipy.stats import ttest_ind

# Filtering the data for t-test
df_drive_1 = df[df['Drive type'] == 1]['Range (WLTP) [km]']
df_drive_2 = df[df['Drive type'] == 2]['Range (WLTP) [km]']

# T-test for Drive Type differences in Range
t_stat, p_value = ttest_ind(df_drive_1, df_drive_2, nan_policy='omit')
print(f"T-statistic: {t_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject null hypothesis: Significant difference in range based on drive type")
else:
    print("Fail to reject null hypothesis: No significant difference in range based on drive type")
```

### Task 8: Inferential Statistics â€“ Hypothesis Testing: Test whether there is a significant difference
in the average Engine power [KM] of vehicles manufactured by two leading manufacturers i.e.
Tesla and Audi. What insights can you draw from the test results?
(Conduct a two sample t-test using ttest_ind from scipy.stats module) (10 Marks)

```python
from scipy.stats import ttest_ind

# Filter the data for Tesla and Audi
tesla_data = df[df['Make'] == 'Tesla']['Engine power [KM]']
audi_data = df[df['Make'] == 'Audi']['Engine power [KM]']

# Perform the two-sample t-test
t_stat, p_value = ttest_ind(tesla_data, audi_data, nan_policy='omit')

# Output the t-statistic and p-value
print(f"T-statistic: {t_stat}, P-value: {p_value}")

# Hypothesis Testing Conclusion
if p_value < 0.05:
    print("Reject null hypothesis: Significant difference in engine power between Tesla and Audi vehicles")
else:
    print("Fail to reject null hypothesis: No significant difference in engine power between Tesla and Audi vehicles")
```

### Task 9: Predictive Insights for Future Trends : Forecast future trends in the electric vehicle market. (10 Marks)

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Check if the dataset contains a year column (assuming it's 'Year')
# If 'Year' column is present, proceed with regression

# Example: Predicting Price based on Year (adjust column names as needed)
if 'Year' in df.columns:
    # Filter relevant columns: 'Year' and 'Price_PLN'
    df_year = df[['Year', 'Minimal price (gross) [PLN]']].dropna()

    # X (Year) and y (Price) for prediction
    X = df_year[['Year']]
    y = df_year['Minimal price (gross) [PLN]']

    # Create and fit the regression model
    model = LinearRegression()
    model.fit(X, y)

    # Predict future trends (e.g., for the next 5 years)
    future_years = np.array([[2025], [2026], [2027], [2028], [2029]])
    predicted_prices = model.predict(future_years)

    # Plotting the prediction
    plt.figure(figsize=(10, 6))
    plt.scatter(df_year['Year'], df_year['Minimal price (gross) [PLN]'], color='blue', label='Historical Data')
    plt.plot(future_years, predicted_prices, color='red', linestyle='dashed', label='Predicted Prices')
    plt.title('Predicted Price of Electric Vehicles Over Time')
    plt.xlabel('Year')
    plt.ylabel('Price (PLN)')
    plt.legend()
    plt.show()

    # Display predictions for future years
    print("Predicted Prices for the next 5 years:")
    for year, price in zip(future_years.flatten(), predicted_prices):
        print(f"Year {year}: {price:.2f} PLN")
else:
    print("No 'Year' column available for time-based forecasting. Consider forecasting based on other features.")
```
